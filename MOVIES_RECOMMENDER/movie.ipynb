# TARGET VECTORIZATION

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=4000,stop_words='english')
```

```python
cv
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
vectors.shape#movie*top 4000words
```

now we will calculate distance between recent movie to all the movies.
***
if distance is large then there is less similarity and vice-versa.
***
but here we will cal angle between them
hence we have to cal similarity of each movie to the every other movies.
***

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity
```

```python
new_df[new_df['title']=='Jeans'].index[0]#finding index of any movie in dataset
```

```python
list(enumerate(similarity[0]))#store the index of movie
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]#top 5 similar movie
```

```python
def recommend(movie):
    ind=new_df[new_df['title']==movie].index[0]
    sim=similarity[ind]
    top_five_movie=sorted(list(enumerate(sim)),reverse=True,key=lambda x:x[1])[1:6]
    for i in top_five_movie:
        print(new_df.iloc[i[0]].title)
```

```python
new_df['title']
```

```python
recommend('Zubaan')
```
