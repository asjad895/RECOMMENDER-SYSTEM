```python
x=rating['user_id'].value_counts()>200#taking users which gives greater than 200 rating
```

```python
y=x[x].index
```

```python
rating=rating[rating['user_id'].isin(y)]
```

```python
rating_with_books=rating.merge(books,on='ISBN')
```

```python
rating_with_books.shape
```

```python
no_of_ratings=rating_with_books.groupby('title')['rating'].count().reset_index().rename(columns={'rating':'no_of_rating'})
```

```python
no_of_ratings
```

```python
final_rating=no_of_ratings.merge(rating_with_books,on='title')
```

```python
final_rating
```

```python
final_rating=final_rating[final_rating['no_of_rating']>50]
```

```python
final_rating
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
title=final_rating.title.unique()
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
nn_model=NearestNeighbors(algorithm='brute')
```

```python
nn_model
```

```python
nn_model.fit(book_sparse)
```

```python
distances,suggestion=nn_model.kneighbors(book_pivot.iloc[507,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distances
```

```python
def recommend(book):
    ind=np.where(book_pivot.index==book)[0][0]
    distances,suggestion=nn_model.kneighbors(book_pivot.iloc[ind,:].values.reshape(1,-1),n_neighbors=6)
    rec_book=[]
    for i in range((len(suggestion))):
        rec_book.append(book_pivot.index[suggestion[i]])
    return rec_book
   
    
    
```

```python
x=recommend('Animal Farm')
```
